```python
import numpy as np
import matplotlib.pyplot as plt

#a = np.sin(np.linspace(1,120,120))
a = np.random.normal(0,10,120)

slope_30 = [0]
for x in range(1, 114, 3):
       slope_30.append(np.mean(a[(x-1):x+2])-np.mean(a[(x+2):(x+5)])) 

print(len(slope_30))

##to color code at different time-point for real data points (object a)
#0-5 minutes "red"
#5-10 minutes "green"
#10-20 minutes "blue"
col_1 = []
for x in range(0,31,1):
    col_1.append("red")

for x in range(31, 61, 1):
    col_1.append("green")
    
for x in range(61,120,1):
    col_1.append("blue")

print("col_1", ":", len(col_1))


##to color code at different time-point for slopes (object a)
#0-5 minutes "red"
#5-10 minutes "green"
#10-20 minutes "blue"
#minimium slope between 5-10 minutes "magenta"

col_2 = []
for x in range(0,11,1):
    col_2.append("red")

for x in range(11, 21, 1):
    col_2.append("green")
    
for x in range(21,39,1):
    col_2.append("blue")

col_2[slope_30.index(min(slope_30[11:21]), 11, 21)] = "magenta"
    
print("col_2", ":", len(col_2))
print(col_2)

plt.figure(1)
##subplot() nrow, ncol, index
plt.subplot(311)
for x in range(0, 120,1):
    plt.scatter(x, a[x], c = col_1[x])
plt.plot(a)
plt.xticks([0,29,59,89,119],["0min", "5min", "10min", "15min", "20min"])
plt.title("Data")

plt.subplot(313)
for i in range(0,39,1):
    plt.scatter(x = list(range(1,117,3))[i], y = slope_30[i], c = col_2[i])

plt.xticks([0,29,59,89,119],["0min", "5min", "10min", "15min", "20min"])
plt.title("Slope")

plt.show()
```
